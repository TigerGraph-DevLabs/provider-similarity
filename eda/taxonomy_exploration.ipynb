```python
def get_super(row):
    try:
        higherDf = df[df["Specialization"].isna()]
        superCode = higherDf[higherDf["Classification"] == row["Classification"]]["Code"].unique()
        return superCode[0]
    except:
        return None

df["Super"] = df[~df["Specialization"].isna()].apply(get_super, axis=1)
```
